```
d = dict() 
from glob import glob
from nltk.corpus import words
filenames = glob('/content/Group15Data/Autogenerated Annotations/TrainOutput/*.csv')
filenames.extend(glob('/content/Group15Data/Autogenerated Annotations/TrainOutput.json'))
filenames.extend(glob('/content/Group15Data/Manual Annotation/TrainOutput/.*csv'))
filenames.extend(glob('/content/Group15Data/Manual Annotation/TrainOutput.json'))
for f in filenames:
    outfile = open(f,'r')
    data = outfile.readlines()
    outfile.close()

    for line in data:
      # Remove the leading spaces and newline character 
      line = line.strip()
      # Convert the characters in line to  
      # lowercase to avoid case mismatch   
      line = line.lower() 
      # Split the line into words 
      wordings = line.split(" ") 
      for word in wordings:
        if word in d:
          d[word] = d[word] + 1 
        else:
          d[word] = 1  

feeds=[]
dic_f = dict()
import nltk
nltk.download('words')

for key in list(d.keys()):
  str_key = str(key)
  if (str_key in words.words()):
    if d[key] < 10 or len(key)==1:  # setting threshold frequency of 10 
      continue
    else:
     dic_f[key] = d[key]
     print(key)
```

```
import json
with open("freq_list.json", "w") as outfile:  
    json.dump(dic_f, outfile, indent = 4) 
```
