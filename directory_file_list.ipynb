Code to generate list of training/dev files in directory. 

```
from os import listdir
from os.path import isfile, join
import json

mypath= "/content/drive/My Drive/Group15Data/Autogenerated Annotations/TrainAOutput"  # enter path to directory here 


l = [f for f in listdir(mypath) if isfile(join(mypath, f))]

out_file = open("TrainAOutput_list.json", "w") 
  
json.dump(l, out_file, indent = 4) 
  
out_file.close() 
```
